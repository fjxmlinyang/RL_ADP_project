```python
# put all together as a function

import matplotlib.pyplot as plt

def total_cost(date):

  df = pd.read_csv('/drive/My Drive/Colab Notebooks/Current_Cost_Total_Results_199_'+ date+'.csv')
  df_1 = df.set_axis(['Scenario','Total_Cost'], axis = 1)
  scen = len(df_1)+1
  bar, ax = plt.subplots()
  ax = sns.lineplot(x='Scenario', y='Total_Cost' ,data = df_1, marker='o')
  ax.set_title('Total_Cost_Abs_Sce_'+str(scen)+'_'+ date)
  ax.set_xlabel('Scenario')
  ax.set_ylabel('Total_Cost')
  plt.xticks(rotation=35) #important!
  #ax.tick_params()
  #ax.legend(loc=4,labelspacing=2,handlelength=3,fontsize=4,shadow=True)

  plt.savefig('Total_Cost_Abs_Sce_'+str(scen)+'_'+ date +'.png', dpi=300)



date_set = ['March 07 2019', 'April 01 2019', 'April 15 2019','April 22 2019']
date_total_cost_true= [17014, 16249, 8294, 18766]
for date in date_set:
  total_cost(date)
```

```python
# put all together as a function

import matplotlib.pyplot as plt

def total_cost_with_true(date, true):
  df = pd.read_csv('/drive/My Drive/Colab Notebooks/Current_Cost_Total_Results_49_'+ date+'.csv')
  df_1 = df.set_axis(['Scenario','Total_Cost'], axis = 1)
 #minus the true value
  df_1['Total_Cost'] = abs((df_1['Total_Cost']-true))/true *100
  scen = len(df_1)+1
  bar, ax = plt.subplots()
  ax = sns.lineplot(x='Scenario', y='Total_Cost' ,data = df_1, marker='o')
  ax.set_title('Error_Rate_Sce_'+str(scen)+'_'+ date)
  ax.set_xlabel('Scenario')
  ax.set_ylabel('Error_Rate_%')
  plt.xticks(rotation=35) #important!
  #ax.tick_params()
  #ax.legend(loc=4,labelspacing=2,handlelength=3,fontsize=4,shadow=True)

  plt.savefig('Error_Rate_Sce_'+str(scen)+'_'+ date +'.png', dpi=300)



date_set = ['March 07 2019', 'April 01 2019', 'April 15 2019','April 22 2019']

date_total_cost_true= [17014, 16249, 8294, 18766]
#how to use this?
for index, date in enumerate(date_set):
  total_cost_with_true(date, date_total_cost_true[index])
```
